### Load library from google colab

```python
import cv2
from google.colab.patches import cv2_imshow
```

### Read and Display the image

```python
image = cv2.imread("mypic.jpg")
cv2_imshow(image)
```

### Load RGB image

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Converting RGB to Grayscale image

```python
# convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

### Inverting Grayscale

```python
# Invert the grayscale image
inverted_gray_image = 255 - gray_image
cv2_imshow(inverted_gray_image)
```

### Applying Gaussian Function

```python
# Apply Gaussian Blur to the inverted image
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
cv2_imshow(blurred_image)
```

### Blending grayscale image

```python
# Blend the grayscale image with the blurred image using the "color dodge" blend mode
def dodgeV2(image, mask):
  return cv2.divide(image,255-mask,scale=256)
pencil_sketch = dodgeV2(gray_image, blurred_image)
cv2_imshow(pencil_sketch)
```

### Original Image vs Pencil Sketch

```python
# Display the original image and pencil sketch
print("RGB image")
cv2_imshow(image)
print("\n")
print("Pencil sketch")
cv2_imshow(pencil_sketch)
```
